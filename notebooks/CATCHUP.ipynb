```python
from cltkreaders.lat import LatinTesseraeCorpusReader
import random
import pandas as pd
from tabulate import tabulate
from pprint import pprint
```

```python
# Set up the corpus reader

CR = LatinTesseraeCorpusReader()
```

```python
# Get fileids

files = CR.fileids()
print(f'There are {len(files)} files in this text collection.')
```

```python
# Subset corpus by metadata

plin = CR.fileids(date='77')
pprint(random.sample(plin, 6))
```

```python
text = next(CR.texts(plin))
print(text[:100])
```

```python
doc = next(CR.spacy_docs(plin))
```

```python
data = []

for token in doc[:5]:
    data.append([token.text, token.lemma_, token.pos_, token.tag_, token.dep_, token.ent_type_])

print(tabulate(data, headers=['Text', 'Lemma', 'POS', 'Tag', 'Dependency', 'Entity Type']))
                
```

```python
# Get sents

sents = list(CR.sents(plin))
for i, sent in enumerate(sents[:5]):
    print(f'Sentence {i}: {sent}')
```

```python
tokens = list(CR.tokens(plin))
for i, token in enumerate(tokens[:5]):
    print(f'Token {i}: {token}')
```

```python
sample_token = next(CR.tokens(plin))
print(sample_token._.citation)
```

```python
import nltk
from nltk import word_tokenize
```

```python
from glob import glob

files = sorted(glob('/Users/katherinethomson/Downloads/*.txt'))

print(len(files))
```

```python
files
```

```python
texts = []

for file in files:
    with open(file, 'r', encoding='utf-8', errors='replace') as f:
        text = f.read()
        texts.append(text)

print(len(texts))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
MIN_DF = 3
MAX_DF = 25
LOWERCASE = True
STOPWORDS = None
TOKEN_PATTERN = r'\b[^\W\d_]+\b'

# cf. https://stackoverflow.com/a/35615151
# min_df is used for removing terms that appear too infrequently. For example:
#     min_df = 0.01 means "ignore terms that appear in less than 1% of the documents".
#     min_df = 5 means "ignore terms that appear in less than 5 documents".
# max_df is used for removing terms that appear too frequently, also known as "corpus-specific stop words". For example:
#     max_df = 0.50 means "ignore terms that appear in more than 50% of the documents".
#     max_df = 25 means "ignore terms that appear in more than 25 documents".

CV = CountVectorizer(input='content', min_df=MIN_DF, max_df=MAX_DF, stop_words=None, lowercase=LOWERCASE, token_pattern=TOKEN_PATTERN)
```

```python
dtm = CV.fit_transform(texts)
```

```python
print(type(dtm))
```

```python
dtm.shape
```

```python
vocab = CV.get_feature_names_out()
```

```python
print(vocab[:10])
```

```python
def filename2label(filename):
    return filename.split('/')[-1].split('.')[0]

labels = [filename2label(file) for file in files]
print(labels)
```
